```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
import torch.optim as optim
import matplotlib.pyplot as plt

class BetaVAE(nn.Module):
    def __init__(self, latent_dim=64, beta=1.0):
        super(BetaVAE, self).__init__()
        self.beta = beta
        self.latent_dim = latent_dim

        self.encoder_conv1 = nn.Conv2d(1, 8, 3, stride=2, padding=1)
        self.encoder_relu1 = nn.ReLU()
        self.encoder_conv2 = nn.Conv2d(8, 16, 2, stride=2, padding=0)
        self.encoder_relu2 = nn.ReLU()
        self.encoder_bn1 = nn.BatchNorm2d(16)
        self.encoder_conv3 = nn.Conv2d(16, 32, 3, stride=2, padding=0)
        self.encoder_relu3 = nn.ReLU()
        self.encoder_flatten = nn.Flatten()
        self.encoder_fc = nn.Linear(32 * 3 * 3, 128)
        self.fc_mu = nn.Linear(128, latent_dim)
        self.fc_logvar = nn.Linear(128, latent_dim)

        self.decoder_fc1 = nn.Linear(latent_dim, 128)
        self.decoder_relu1 = nn.ReLU()
        self.decoder_fc2 = nn.Linear(128, 32 * 3 * 3)
        self.decoder_relu2 = nn.ReLU()
        self.decoder_deconv1 = nn.ConvTranspose2d(32, 16, 3, stride=2, padding=0, output_padding=0)
        self.decoder_relu3 = nn.ReLU()
        self.decoder_bn1 = nn.BatchNorm2d(16)
        self.decoder_deconv2 = nn.ConvTranspose2d(16, 8, 2, stride=2, padding=0, output_padding=0)
        self.decoder_relu4 = nn.ReLU()
        self.decoder_deconv3 = nn.ConvTranspose2d(8, 1, 3, stride=2, padding=1, output_padding=1)
        self.decoder_sigmoid = nn.Sigmoid()

    def encode(self, x):
        x = self.encoder_conv1(x)
        x = self.encoder_relu1(x)
        x = self.encoder_conv2(x)
        x = self.encoder_relu2(x)
        x = self.encoder_bn1(x)
        x = self.encoder_conv3(x)
        x = self.encoder_relu3(x)
        x = self.encoder_flatten(x)
        x = self.encoder_fc(x)
        mu = self.fc_mu(x)
        logvar = self.fc_logvar(x)
        return mu, logvar

    def reparameterize(self, mu, logvar):
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return mu + eps * std

    def decode(self, z):
        x = self.decoder_fc1(z)
        x = self.decoder_relu1(x)
        x = self.decoder_fc2(x)
        x = self.decoder_relu2(x)
        x = x.view(-1, 32, 3, 3)
        x = self.decoder_deconv1(x)
        x = self.decoder_relu3(x)
        x = self.decoder_bn1(x)
        x = self.decoder_deconv2(x)
        x = self.decoder_relu4(x)
        x = self.decoder_deconv3(x)
        recon_x = self.decoder_sigmoid(x)
        return recon_x

    def forward(self, x):
        mu, logvar = self.encode(x)
        z = self.reparameterize(mu, logvar)
        recon_x = self.decode(z)
        return recon_x, mu, logvar

    def loss_function(self, recon_x, x, mu, logvar):
        BCE = F.binary_cross_entropy(recon_x, x, reduction='sum')
        KLD = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp())
        loss = BCE + self.beta * KLD
        return loss

def train(model, device, train_loader, optimizer, epoch, log_interval=100):
    model.train()
    train_loss = 0
    for batch_idx, (data, _) in enumerate(train_loader):
        data = data.to(device)
        optimizer.zero_grad()
        recon_batch, mu, logvar = model(data)
        loss = model.loss_function(recon_batch, data, mu, logvar)
        loss.backward()
        train_loss += loss.item()
        optimizer.step()

        if batch_idx % log_interval == 0:
            print(f'Train Epoch: {epoch} [{batch_idx * len(data)}/{len(train_loader.dataset)} '
                  f'({100. * batch_idx / len(train_loader):.0f}%)]\tLoss: {loss.item() / len(data):.6f}')

    average_loss = train_loss / len(train_loader.dataset)
    print(f'====> Epoch: {epoch} Average loss: {average_loss:.4f}')

def test(model, device, test_loader):
    model.eval()
    test_loss = 0
    with torch.no_grad():
        for data, _ in test_loader:
            data = data.to(device)
            recon, mu, logvar = model(data)
            loss = model.loss_function(recon, data, mu, logvar)
            test_loss += loss.item()

    average_loss = test_loss / len(test_loader.dataset)
    print(f'====> Test set loss: {average_loss:.4f}')

def visualize_reconstructions(model, device, test_loader, num_images=8):
    model.eval()
    with torch.no_grad():
        data, _ = next(iter(test_loader))
        data = data.to(device)
        recon, _, _ = model(data)
        data = data.cpu()
        recon = recon.cpu()

        fig, axes = plt.subplots(2, num_images, figsize=(num_images * 2, 4))
        for i in range(num_images):
            axes[0, i].imshow(data[i].squeeze(), cmap='gray')
            axes[0, i].axis('off')
            axes[1, i].imshow(recon[i].squeeze(), cmap='gray')
            axes[1, i].axis('off')
        plt.show()

latent_dim = 10
beta = 4.0
batch_size = 128
learning_rate = 1e-3
num_epochs = 20
log_interval = 100

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f'Using device: {device}')

transform = transforms.Compose([
    transforms.ToTensor(),
])

train_dataset = datasets.FashionMNIST(
    root='.', 
    train=True, 
    transform=transform, 
    download=True
)

test_dataset = datasets.FashionMNIST(
    root='.', 
    train=False, 
    transform=transform, 
    download=True
)

train_loader = DataLoader(
    dataset=train_dataset, 
    batch_size=batch_size, 
    shuffle=True
)

test_loader = DataLoader(
    dataset=test_dataset, 
    batch_size=batch_size, 
    shuffle=False
)

model = BetaVAE(latent_dim=latent_dim, beta=beta).to(device)
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

for epoch in range(1, num_epochs + 1):
    train(model, device, train_loader, optimizer, epoch, log_interval)
    test(model, device, test_loader)
    if epoch % 5 == 0:
        visualize_reconstructions(model, device, test_loader)

visualize_reconstructions(model, device, test_loader)
torch.save(model.state_dict(), 'beta_vae_fashion_mnist.pth')
print('Model saved as beta_vae_fashion_mnist.pth')
```

# How we choose beta
The value \( \beta = 4.0 \) was chosen to increase the regularization strength of the latent space in the Beta-VAE model. A higher \( \beta \) encourages more disentanglement of the latent variables by giving more weight to the KL divergence term. This helps to better structure the latent space but may slightly reduce reconstruction quality. It's a reasonable starting point to balance both reconstruction accuracy and the regularization of the latent space.

The **ConvTranspose2d** layer in neural networks is used for **upsampling** and increases the spatial dimensions of the data.

### Formula for Calculating Output Dimensions:
`Output_size = (Input_size - 1) * Stride - 2 * Padding + Kernel_size + Output_padding`

### Effect of Layer Parameters:
- **Kernel size**: Determines how much data is processed at each step.
- **Stride**: Determines the extent of spatial dimension expansion.
- **Padding** and **Output padding**: Adjust the final output dimensions.

### Example:
- **First layer**: `(6 - 1) * 2 + 3 = 13`
- **Second layer**: `(13 - 1) * 1 + 2 = 14`
- **Third layer**: `(14 - 1) * 2 - 2 * 1 + 3 + 1 = 28`

```python
import matplotlib.pyplot as plt
import torch
import math

def visualize_reconstructions_and_random_samples(model, device, test_loader, num_samples=100):
    """
    Visualizes reconstructed test samples and generated random samples in a grid.

    Args:
        model: Trained VAE model.
        device: Device (CPU or GPU).
        test_loader: DataLoader for the test dataset.
        num_samples: Number of samples to visualize.
    """
    model.eval()

    # Get a batch of test samples
    with torch.no_grad():
        test_data, _ = next(iter(test_loader))
        test_data = test_data[:num_samples].to(device)  # Select the first `num_samples`
        reconstructed, _, _ = model(test_data)

        # Generate random samples from the latent space
        latent_dim = model.latent_dim
        random_latents = torch.randn(num_samples, latent_dim).to(device)
        random_samples = model.decode(random_latents)

    # Move tensors to CPU for visualization
    test_data = test_data.cpu()
    reconstructed = reconstructed.cpu()
    random_samples = random_samples.cpu()

    # Define grid size based on number of samples
    grid_size = int(math.ceil(math.sqrt(num_samples)))  # We aim for a square grid

    # Plot reconstructed test samples
    fig, axes = plt.subplots(grid_size, grid_size, figsize=(grid_size * 2, grid_size * 2))
    axes = axes.flatten()  # Flatten the axes to easily iterate over them

    for i in range(num_samples):
        axes[i].imshow(test_data[i].squeeze(), cmap="gray")
        axes[i].axis("off")
    for i in range(num_samples, len(axes)):
        axes[i].axis("off")  # Hide unused subplots
    plt.suptitle("Original Test Samples", y=1.05)
    plt.show()

    # Plot reconstructed test samples
    fig, axes = plt.subplots(grid_size, grid_size, figsize=(grid_size * 2, grid_size * 2))
    axes = axes.flatten()

    for i in range(num_samples):
        axes[i].imshow(reconstructed[i].squeeze(), cmap="gray")
        axes[i].axis("off")
    for i in range(num_samples, len(axes)):
        axes[i].axis("off")  # Hide unused subplots
    plt.suptitle("Reconstructed Test Samples", y=1.05)
    plt.show()

    # Plot random samples generated from the latent space
    fig, axes = plt.subplots(grid_size, grid_size, figsize=(grid_size * 2, grid_size * 2))
    axes = axes.flatten()

    for i in range(num_samples):
        axes[i].imshow(random_samples[i].squeeze(), cmap="gray")
        axes[i].axis("off")
    for i in range(num_samples, len(axes)):
        axes[i].axis("off")  # Hide unused subplots
    plt.suptitle("Generated Random Samples", y=1.05)
    plt.show()

visualize_reconstructions_and_random_samples(model, device, test_loader, num_samples=100)
```

### Generation from Noise

When generating images from noise, the resulting images tend to be very similar to one another and lack diversity. Additionally, some generated images contain errors that do not correspond to any recognizable category. While the model is capable of producing samples for each category, the variations within each category are insufficient, leading to a limited range of outputs.

FID (Fréchet Inception Distance) evaluates the quality of generated images by comparing them to real ones. It works by passing both sets of images through a pre-trained model to extract features. Then, it calculates the difference in their distributions (mean and covariance) to give a score: lower is better. For VAE, FID helps measure how realistic and diverse the generated images are compared to the real dataset.
The computed FID scores will give you a sense of:
How close reconstructions are to the test distribution (FID real vs recon)
How the model performs when generating new samples from perturbed latent codes (FID real vs noisy-generated)

```python
import os
import shutil
import torch
import torchvision.utils as vutils
from torch_fidelity import calculate_metrics

# Directories for saving images
real_dir = './fid_real'
recon_dir = './fid_recon'
gen_dir = './fid_gen'

for d in [real_dir, recon_dir, gen_dir]:
    if os.path.exists(d):
        shutil.rmtree(d)
    os.makedirs(d)

def save_images(images, folder, start_idx=0):
    images = images.cpu()
    for i in range(images.size(0)):
        img_path = os.path.join(folder, f'{start_idx+i}.png')
        vutils.save_image(images[i], img_path)

model.eval()
real_images_collected = 0
recon_images_collected = 0

with torch.no_grad():
    for batch_idx, (data, _) in enumerate(test_loader):
        data = data.to(device)
        recon, _, _ = model(data)
        
        # Save real images
        save_images(data, real_dir, start_idx=real_images_collected)
        real_images_collected += data.size(0)
        
        # Save reconstructed images
        save_images(recon, recon_dir, start_idx=recon_images_collected)
        recon_images_collected += recon.size(0)

# 2. Compute FID between real test images and reconstructed images
metrics_recon = calculate_metrics(
    input1=real_dir,
    input2=recon_dir,
    cuda=True if torch.cuda.is_available() else False,
    isc=False,
    fid=True,
    kid=False
)
print("FID (Real vs Recon):", metrics_recon['frechet_inception_distance'])

# 3. Generate new samples by adding noise in latent space or using noisy test data
gen_images_collected = 0
with torch.no_grad():
    for batch_idx, (data, _) in enumerate(test_loader):
        data = data.to(device)
        mu, logvar = model.encode(data)
        z = model.reparameterize(mu, logvar)
        
        noise = torch.randn_like(z) * 0.5  
        z_noisy = z + noise

        gen = model.decode(z_noisy)
        save_images(gen, gen_dir, start_idx=gen_images_collected)
        gen_images_collected += gen.size(0)

# 4. Compute FID between real test images and these newly generated images
metrics_gen = calculate_metrics(
    input1=real_dir,
    input2=gen_dir,
    cuda=True if torch.cuda.is_available() else False,
    isc=False,
    fid=True,
    kid=False
)
print("FID (Real vs Noisy-Generated):", metrics_gen['frechet_inception_distance'])

```


When evaluating Variational Autoencoders (VAEs), there are several alternatives to the Fréchet Inception Distance (FID) . While FID is popular—especially for GANs—VAEs often benefit from other measures too:

1. **Log Likelihood or ELBO**:  
   The Evidence Lower Bound (ELBO) is a classic metric for VAEs since it’s basically a lower bound on the log likelihood of real data under the model . A higher ELBO suggests the VAE is better at capturing the data distribution, though it doesn’t always mean the images "look" better.

2. **Inception Score (IS)**:  
   The Inception Score  judges how "sharp and diverse" generated images are based on a pretrained classifier. It’s often mentioned in the same breath as FID but is slightly older and sometimes less stable.

3. **MSE or PSNR**:  
   Simple reconstruction metrics like Mean Squared Error (MSE) or Peak Signal-to-Noise Ratio (PSNR)  tell you how well a VAE can rebuild the original image. They’re easy to compute but might miss the "perceptual" quality that humans care about.

4. **LPIPS (Perceptual Metrics)**:  
   Learned Perceptual Image Patch Similarity (LPIPS)  tries to mimic human judgments of image similarity. It’s often better at capturing subtle differences in quality than raw MSE. Though it’s not a perfect measure, it can be more "human-friendly."

**Comparing to FID**:  
FID measures how the distribution of generated images differs from that of real images, using features from a pretrained network . It somewhat balances realism and diversity. If you rely only on FID, you might miss other aspects—like how well the VAE represents the underlying data probabilities or how good its reconstructions are. Sometimes I find it just too sensitive.

In practice, people often combine ELBO with something like FID or LPIPS. This gives a more rounded evaluation: ELBO for how well the VAE "understands" the data statistically, and something like LPIPS or FID for how real or appealing the images look.
